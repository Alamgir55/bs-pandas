```python
import numpy as np
```

```python
def convert(day,month,year):
    return f"{day}--{month}--{year}"
```

```python
hotels['date'] = np.vectorize(convert)(hotels['arrival_date_day_of_month'],
                                      hotels['arrival_date_month'],
                                      hotels['arrival_date_year'])
```
